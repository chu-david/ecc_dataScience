In this little program, we are visualising "Countries of the world"

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("assets/countries of the world.csv", encoding='latin-1')

df.head() #preview 
```

df = dataframe

object -> String

Population -> integer (whole number)

```python
df.dtypes

```

```python
barGraph = sns.barplot(x="Region",y="Population",data=df,errorbar = ['ci',0])

barGraph.set_xticklabels(barGraph.get_xticklabels(),
rotation=90,ha='right')
plt.ylabel("population")
plt.xlabel("Regions")
plt.show()
```

```python
boxGraph = sns.boxplot(x="Region",y="Population",data=df)

boxGraph.set_xticklabels(boxGraph.get_xticklabels(),
rotation=90,ha='right')
plt.ylabel("population")
plt.xlabel("countries")
plt.show()
```

```python
import math

histGraph = sns.histplot(y="Population",data=df,binwidth=5*math.pow(10,6))

histGraph.set_xticklabels(histGraph.get_xticklabels(),
rotation=90,ha='right')
plt.ylabel("population")
plt.xlabel("countries")

plt.show()
```
